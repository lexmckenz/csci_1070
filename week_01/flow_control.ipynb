# Loops and Functions

```python
# loops are flow control. They change how the code 
# executes iteratively or how code or data moves through
# the application 
# if ... (elif ... else)
# elif and else are optional. Depends on what youre trying
# to do if you need to use them. BUT it is best practice
# to at least use an else

# if a equals b, return "they are equal"
# if a is greater than b, return "a is greater"
# if be is greater than a, return "b is greater"

# == is comparing values whereas = is setting something
# equal to something else

a = 20 
b = 13

if a == b:
    print("they are equal")
elif a>b:
    print("a is greater")
# else:
elif b>a:
    print("b is greater")

# pseudocode:
# if a equals b:
#   print "they are equal"
# else if a>b:
#   print "a is greater"
# else if b>a:
#   print "b is greater"
```

```python
# while loops
# repeatedly tests an expression and continues to run whatever 
# is in the loop until a stopping conditon is met
x = 7
while x > 5:
    print(x)
    #decrement our x so it can stop
    # x = x - 1
    x-=1
```

```python
# range(start, stop[, step])
# stop is exclusive, so its not included
for r in range(1,10,2):
    print(r)
```

```python
# if we our code to be reusable, we use a function!
# and variables we want to use in the function need to be passed in
# or defined inside the function itself.
def comparison(a=0,b=0):
    if a == b: 
        return "they are equal"
    elif a>b:
        return "a is greater"
    elif b>a: 
        return "b is greater"
```
