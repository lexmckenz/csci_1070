# Constructors 

Constructors get created when you create an instance of a class. 

__init__(__ is "dunder")

SELF is a default variable that contains the memory address of the current you're working in.

```python
class a_sample_class:
    #default/basic constructor
    def __init__(self):
        self.var = 0

    # an instance method for printing
    def print_vars(self):
        print(self.var)

# create instance of the class
obj = a_sample_class()

obj.print_vars()
```

```python
class a_sample_class:
    #parameterized constructor
    def __init__(self, one, two):
        self.first = one
        self.second = two

    # an instance method for printing
    def print_vars(self):
        print(f"first: {self.first}, second: {self.second}")

# create instance of the class
obj = a_sample_class(7,43)

obj.print_vars()
```

# Instance, Class, and Static Methods

```python
class Candy: 
    #a reason we use dunder is because 
    # single underscores are reserve for internal purposes
    def __init__(self, _brand,
                 _is_caramel=False,
                 _is_chocolate=False,
                 _is_vegan=False,
                 _has_nuts=False,
                 _type="hard candy",
                 _calories=200,
                 _size="funsized"):
        self.brand = _brand
        self.is_caramel=_is_caramel
        self.is_chocolate=_is_chocolate
        self.is_vegan=_is_vegan
        self.calories=_calories
        self.type=_type

    #class method - used to modify the state of the
    #class (for all instances)
    @classmethod
    def snickers(cls):
        return cls('Mars', _is_caramel=True, _has_nuts=True, _type="Snickers")
    
    #instance method - can access and modify both the
    #class and instance state
    #printing is accessing state
    def display(self):
        print(f"My favorite candy is {self.brand} {self.type}")

    #static method - used for comparison or validation
    #commonly used in data for checking if data meets
    #particular conditons 
    # used if you want to check the value without needing
    #an instance of it
    @staticmethod
    def calories(size, n):
        if size == "large":
            calories = 500
        elif size == "regular":
            calories = 400
        elif size == "funsized":
            calories = 100
        else:
            calories = 1000
        
        calories = calories*n

        return calories
```

```python
snickers_bar = Candy.snickers()

snickers_bar.display()
```

```python
favorite_candy = Candy(_brand="Mars",_type="Skittles")
favorite_candy.display()
```
