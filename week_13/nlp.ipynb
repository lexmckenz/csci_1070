# Stop Words

```python
# stop words are common words that have
# low information value.
# nltk.download('stopwords')
sw = set(nltk.corpus.stopwords.words('english'))
sw
```

```python
text_sentences = nltk.corpus.gutenberg.sents('carroll-alice.txt')[:10]
text_sentences
```

```python
# stop words are all lowercase 
# so we want to make sure all our words are
# lowercase when the process them
for sentence in text_sentences: 
    filtered_list = [word for word in sentence if word.lower() not in sw]
    print(filtered_list)
```

# Stemming

```python
# stemming - finding root words
# fun fact: Google adopted word stemming in 2003
# it still does this but it uses a more sophisticated
# approach. Before this, search words like "fish"
# would not give results for "fishing" or "fishes"
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer(language='english')
word = 'changing'
stemmer.stem(word)
```

# Lemmatization

```python
# lemmatization - finding the form of a word thats related
# in the dictionary. The process for calculating "lemmas"
# is more complex than stemming but more accurate
# lemmatization considers context AND convert the word
# to its "meaninfgul base form" (a lemma). Words
# can have mutliple lemmas
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
word = 'dances'
lemmatizer.lemmatize(word)
```

# Sentiment Analysis

```python
# nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
text = """
        AWESOME
        """
```

```python
sentiment = analyzer.polarity_scores(text)
print(sentiment)
```

```python
import pandas as pd
sentiments = []
compound = sentiment['compound']
pos = sentiment['pos']
neg = sentiment['neg']
neu = sentiment['neu']

sentiments.append({
    "text": text,
    "compound": compound,
    "positive": pos,
    "negative": neg,
    "neutral": neu
})

pd.DataFrame(sentiments)
```
